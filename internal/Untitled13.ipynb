```python
from ultralytics import YOLO
import cv2

# Load YOLO model
model = YOLO("yolov8m.pt")

# Predict
results = model.predict("cat_dog.jpg")

# Get the first element of the results list
result = results[0]

# Load the image
image = cv2.imread("cat_dog.jpg")

# Loop through detected boxes and draw them on the image
for box in result.boxes:
    class_id = box.cls[0].item()
    class_name = result.names[class_id]
    confidence = box.conf[0].item()
    xmin, ymin, xmax, ymax = map(int, box.xyxy[0].tolist())

    # Draw box with label
    cv2.rectangle(image, (xmin, ymin), (xmax, ymax), (0, 255, 0), 2)
    cv2.putText(image, f"{class_name}: {confidence:.2f}", (xmin, ymin - 10),
                cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 255, 0), 2)

# Show the image with detected objects
cv2.imshow("Detected Objects", image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
